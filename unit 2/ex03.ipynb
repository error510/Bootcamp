```python
#outliers 
df['NAsales_zscore']=(df.NA_Sales-df.NA_Sales.mean() ) / df.NA_Sales.std()
df['EUsales_zscore']=(df.EU_Sales-df.EU_Sales.mean() ) / df.EU_Sales.std()
df['JPsales_zscore']=(df.JP_Sales-df.JP_Sales.mean() ) / df.JP_Sales.std()
df['Globalsales_zscore']=(df.Global_Sales-df.Global_Sales.mean() ) / df.Global_Sales.std()
df['Rank_zscore']=(df.Rank-df.Rank.mean() ) / df.Rank.std()
df['Year_zscore']=(df.Year-df.Year.mean() ) / df.Year.std()
pre_df = df 
df = df[(abs(df['Rank_zscore'])>1 )
     &(df['Rank_zscore'] < df['Rank_zscore'].max()) 
     & (abs(df['Year_zscore'])> 1) 
     &  (abs(df['Globalsales_zscore'])>1)
     & (abs(df['JPsales_zscore'])> 1)
     & (abs(df['NAsales_zscore'])> 1) 
     & (abs(df['EUsales_zscore'])> 1)
     & (df['Rank_zscore'] < df['Rank_zscore'].max()) 
     & (df['NAsales_zscore'] < df['NAsales_zscore'].max()) 
     & (df['EUsales_zscore'] < df['EUsales_zscore'].max())
     &(df['JPsales_zscore'] < df['JPsales_zscore'].max())
     &(df['Globalsales_zscore'] < df['Globalsales_zscore'].max()) 
     &(df['Year_zscore'] < df['Year_zscore'].max())]
```
