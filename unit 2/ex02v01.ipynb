```python
import pandas as pd 
from matplotlib import pyplot 
import math 
df = pd.read_csv('./vgsales.csv')
#calculating the requested quantities 
euSales = df.EU_Sales
theMean = euSales.mean()
theMedian = euSales.median()
theMode = euSales.mode()
thaVariance = euSales.var()
theSD = math.sqrt(thaVariance)
theSD2 = euSales.std()
#ploting the 3 figures
df.dropna(inplace=True)
new_df = df.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64'])
new_df = new_df.fillna(new_df.median())
#scatter plot
x = new_df.Rank 
y = new_df.EU_Sales
pyplot.scatter(x,y,5)
pyplot.show()
new_df.plot.scatter(x='Rank', y='EU_Sales')
#histogram
pyplot.hist(y,15)
pyplot.show()
y.hist(bins=10)
#box plot 
pyplot.boxplot(y , widths = 10 )
new_df.boxplot('EU_Sales')
```
