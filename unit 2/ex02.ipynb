```python
import pandas as pd 
from matplotlib import * 
import math 
df = pd.read_csv('./vgsales.csv')
#calculating the requested quantities 
euSales = df.EU_Sales
theMean = euSales.mean()
theMedian = euSales.median()
theMode = euSales.mode()
thaVariance = euSales.var()
theSD = math.sqrt(thaVariance)
theSD2 = euSales.std()
#ploting the 3 figures 
#scatter plot
x = df.Rank 
y = df.EU_Sales
matplotlib.pyplot.scatter(x,y,5)
matplotlib.pyplot.show()
df.plot.scatter(x='Rank', y='EU_Sales')
#histogram
matplotlib.pyplot.hist(y,15)
matplotlib.pyplot.show()
y.hist(bins=10)
#box plot 
matplotlib.pyplot.boxplot(y , widths = 10 )
df.boxplot('EU_Sales')
```
